```python
# Merge Orders with Employees using EmployeeID
orders_employees = pd.merge(orders, employees, how='left', on='EmployeeID')

# Merge Products with Categories
products_categories = pd.merge(products, categories, how='left', on='CategoryID')

# Merge Order_Details with Products
order_details_products = pd.merge(order_details, products_categories, how='left', on='ProductID')

# Merge Orders_Customers with Order_Details_Products
orders_customers = pd.merge(orders_employees, customers, how='left', on='CustomerID')

# Merge all the data together (Orders + Order Details + Products + Categories + Customers + Employees)
full_data = pd.merge(orders_customers, order_details_products, how='left', on='OrderID')
```

```python
full_data.isnull().sum()
```

```python
full_data['ShippedDate'].fillna('Unknown', inplace=True)
full_data['ShipRegion'].fillna('Unknown', inplace=True)
full_data['ShipPostalCode'].fillna('Unknown', inplace=True)
full_data['Region_x'].fillna('Unknown', inplace=True)
full_data['ReportsTo'].fillna('Unknown', inplace=True)

full_data['Region_y'].fillna('Unknown', inplace=True)
full_data['PostalCode_y'].fillna('Unknown', inplace=True)
full_data['Fax'].fillna('Unknown', inplace=True)
full_data
```

```python
full_data.isnull().sum()
```

```python
# Add a column for revenue (UnitPrice * Quantity)
full_data['Revenue'] = full_data['UnitPrice_y'] * full_data['Quantity']

# Group by Product and Category to get total revenue and quantity sold
product_sales = full_data.groupby(['ProductName', 'CategoryName'])['Revenue'].sum().reset_index()

# Sort by Revenue to identify top-selling products
top_products = product_sales.sort_values(by='Revenue', ascending=False).head(10)

# Display top-selling products
print("Top selling products:")
print(top_products)
```

```python
# Plot top-selling products by revenue
plt.figure(figsize=(12, 8))
top_products_plot = sns.barplot(x='Revenue', y='ProductName', data=top_products, palette='viridis')
top_products_plot.set_title('Top 10 Selling Products by Revenue', fontsize=16)
top_products_plot.set_xlabel('Revenue ($)', fontsize=14)
top_products_plot.set_ylabel('Product', fontsize=14)
plt.show()
```

## Thüringer Rostbratwurst and Raclette Courdavault are the second and third highest revenue generators, with revenues slightly above 100,000.

## Insight: There is a clear disparity between the top-selling product (Côte de Blaye) and others. The revenue gap suggests that this product either has premium pricing or is a market favorite.

```python
# Convert OrderDate and ShippedDate to datetime format
full_data['OrderDate'] = pd.to_datetime(full_data['OrderDate'])

# Group by YearMonth to get monthly revenue
seasonal_sales = full_data.groupby('OrderDate')['Revenue'].sum().reset_index()

# Plot seasonal sales trends
plt.figure(figsize=(10, 5))
sns.lineplot(x='OrderDate', y='Revenue', data=seasonal_sales)
plt.title('Seasonal Sales Trend')
plt.xlabel('Years and Months')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.show()
```

# The data spans from July 1996 to approximately April 1998.
# 1996:The revenue starts relatively low with minor fluctuations.
# 1997:Noticeable spikes occur in October and early 1997.Revenue shows periodic increases followed by drops, indicating irregular sales cycles.
# 1998:A spike is observed in early 1998 (April), where revenue exceeds $25,000. This could signify a major event like a product launch maybe.

```python
# Group by Region and CustomerID to get total revenue per region
region_sales = full_data.groupby(['Region_x', 'CustomerID'])['Revenue'].sum().reset_index()

# Sort by revenue within each region
region_sales_sorted = region_sales.sort_values(by='Revenue', ascending=False)

# Display the top regions with high sales
top_regions = region_sales_sorted.groupby('Region_x').head(5)
print(top_regions)
```

```python
# Get total revenue per customer and top customers by revenue
customer_sales = full_data.groupby('CustomerID')['Revenue'].sum().reset_index()
high_value_customers = customer_sales.sort_values(by='Revenue', ascending=False).head(10)

# Visualize high-value customers
plt.figure(figsize=(12, 8))
sns.barplot(x='Revenue', y='CustomerID', data=high_value_customers, palette='plasma')
plt.title('Top 10 High-Value Customers by Revenue', fontsize=16)
plt.xlabel('Total Revenue ($)', fontsize=14)
plt.ylabel('Customer ID', fontsize=14)
plt.show()
```

```python
# Get total revenue per customer and top customers by revenue
customer_sales = full_data.groupby('Region_y')['Revenue'].sum().reset_index()
high_value_customers = customer_sales.sort_values(by='Revenue', ascending=False).head(10)

# Visualize high-value customers
plt.figure(figsize=(10, 8))
sns.barplot(x='Revenue', y='Region_y', data=high_value_customers, palette='plasma')
plt.title('Top 10 High-Value Regions by Revenue')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Regions')
plt.show()
```

```python
# Group by EmployeeID to calculate total sales for each employee
employee_sales = full_data.groupby('EmployeeID')['Revenue'].sum().reset_index()

# Sort by total revenue to identify top and bottom employees
employee_sales_sorted = employee_sales.sort_values(by='Revenue', ascending=False)

# Display top and bottom employees by revenue
#top_bottom_employees = pd.concat([employee_sales_sorted.head(5), employee_sales_sorted.tail(5)])
#top_bottom_employees
top_employees=employee_sales_sorted.head()
bottom_employees=employee_sales_sorted.tail(5)
print(top_employees)
print(bottom_employees)
```

```python
# Merge employee data with the full data to get employee names
employee_sales = pd.merge(employee_sales, employees[['EmployeeID', 'FirstName', 'LastName']], on='EmployeeID', how='left')

# Combine first and last names to get a full name
employee_sales['EmployeeName'] = employee_sales['FirstName'] + ' ' + employee_sales['LastName']

# Step 3: Visualizing Employee Sales Performance with Names
plt.figure(figsize=(10, 5))
sns.barplot(x='Revenue', y='EmployeeName', data=employee_sales.sort_values(by='Revenue', ascending=False), palette='Greens')

# Add labels and title
plt.title('Employee-wise Sales Performance (Total Revenue)', fontsize=16)
plt.xlabel('Total Revenue ($)', fontsize=14)
plt.ylabel('Employee Name', fontsize=14)

# Show the plot
plt.show()
```
